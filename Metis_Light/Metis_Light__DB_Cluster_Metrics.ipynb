```python
import configparser

config = configparser.ConfigParser() 
config.read_file(open(r'../ipynb.cfg'))
con_str_metis = config.get('con_str', 'PG_TEMBO_DEV1_METIS') 


%sql {con_str_metis}
```

```python
%%sql 

SELECT count(*) AS connt_row, 
	max(time) AS last_insert,
	date_trunc('second', now() - max(time) )AS time_since_last_insert,
	pg_size_pretty(pg_total_relation_size('metis.pg_stat_database_history')) AS table_size
FROM metis.pg_stat_database_history;
```

```python
%%sql
-- History of the last 7 days
SELECT * 
FROM v_db_activity_last_7d;
```

```python
%%sql 
-- for QA 

WITH db_history_diff AS (
    SELECT datname, 
           tup_updated, 
           LAG(tup_updated) OVER (PARTITION BY datname ORDER BY time) AS previous_value,
           tup_updated - LAG(tup_updated) OVER (PARTITION BY datname ORDER BY time) AS diff_tup_updated,
           date_trunc('second', time) AS time
    FROM metis.pg_stat_database_history
    WHERE datname = 'demo'
      AND time > now() - INTERVAL '30 days'
)
SELECT *
FROM db_history_diff
WHERE diff_tup_updated >= 0
ORDER BY time DESC
LIMIT 10000;
```
