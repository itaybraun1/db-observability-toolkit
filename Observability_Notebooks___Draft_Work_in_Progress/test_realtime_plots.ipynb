```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, clear_output
import time

# Initialize an empty plot
plt.figure(figsize=(10, 6))
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Real-Time Chart Demo')
plt.grid(True)

# Define a function to update the plot with random data
def update_plot():
    x = np.arange(0, 10, 0.1)  # Time values (0 to 10 in increments of 0.1)
    y = np.random.rand(len(x))  # Random data points
    
    plt.clf()  # Clear the existing plot
    plt.plot(x, y, label='Random Data')
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.title('Real-Time Chart Demo')
    plt.grid(True)
    plt.legend()
    plt.xlim(0, 10)
    plt.ylim(0, 1)
    
    # Display the updated plot
    display(plt.gcf())
    clear_output(wait=True)

# Continuously update the plot every 1 second
while True:
    update_plot()
    time.sleep(1)
```

```python
import plotly.graph_objs as go
from IPython.display import display, clear_output
import time
import random

# Initialize an empty figure
fig = go.FigureWidget()
fig.update_layout(
    title="Real-Time Chart Demo",
    xaxis_title="Time",
    yaxis_title="Value"
)
display(fig)

# Initialize empty lists for x and y data
x_data = []
y_data = []

# Define a function to update the plot with random data
def update_plot():
    x_data.append(len(x_data))  # Increment x
    y_data.append(random.random())  # Generate random data point
    
    # Limit the data to the last 10 points for a scrolling effect
    if len(x_data) > 10:
        x_data.pop(0)
        y_data.pop(0)
    
    fig.data = []  # Clear existing data traces
    fig.add_trace(go.Scatter(x=x_data, y=y_data, mode="lines"))
    
    clear_output(wait=True)
    display(fig)

# Continuously update the plot every 1 second
while True:
    update_plot()
    time.sleep(1)
```
